```python
## Select genes identified as highly variable genes (HVGs) in at least one 'mini_set' dataset.
def miniset_genes(gene_list, min_set):
    gene_count = {}
    print(f'min_set: {min_set}')

    gene_sets = [set(_) for _ in gene_list]
    for gene_set in gene_sets:
        for gene in gene_set:
            gene_count[gene] = gene_count.get(gene, 0) + 1

    miniset_gene_list = [gene for gene, count in gene_count.items() if count >= min_set]
    sl_gene_len = len(miniset_gene_list)
    print(f'overlapped gene len: {sl_gene_len}')
    return miniset_gene_list
```

## Vis credible gene set with HVGs of datasets

```python
def overlap_gene_num_single(adata_list, genes_overall, gene_mode = True, gene_col = None, figsize = (18, 1)):
    if not gene_mode:
        if gene_col is None:
            gene_lists = [adata.var_names.tolist() for adata in adata_list]
        else:
            gene_lists = [adata.var[gene_col].tolist() for adata in adata_list]
    else:
        gene_lists = adata_list
    overlap_counts = np.zeros(len(gene_lists))

    # Find the overlap between each pair of adata objects
    for i in range(len(gene_lists)):
        overlap = len(set(genes_overall).intersection(gene_lists[i]))
        overlap_counts[i] = overlap


    # Plot the overlap counts as a heatmap
    fig, ax = plt.subplots(1,1, figsize = figsize)
    sns.heatmap(overlap_counts[None,:], annot=True, fmt="g", cmap="crest")
    plt.xticks(np.arange(len(adata_list)) + 0.5, file_names, rotation=90)
    plt.yticks(range(1), ['genes_overall'], rotation=0)
    # plt.xlabel("Adata Index")
    # plt.ylabel("Adata Index")
    plt.title("Overlap of Genes between Adata Objects")
    plt.show()
```

```python
ct_col = 'ann_finest_level'

def filter_less_ct(df):
    grouped = df.groupby('data_file_names')
    
    filtered_df = pd.DataFrame(columns=[ct_col, 'predictions', 'data_file_names', 'methods'])
    for name, group in grouped:
        counts = group['ann_finest_level'].value_counts()

        valid_cell_types = counts[counts >= 25].index

        filtered_group = group[group['ann_finest_level'].isin(valid_cell_types)]
        filtered_df = pd.concat([filtered_df, filtered_group])

    return filtered_df


adata = adata_inte_mapped.copy()
print(f'before remove nan, adata shape: {adata.shape}')
index = [_ is not np.nan for _ in adata.obs['ann_finest_level']]
adata = adata[index]
print(f'after remove nan, adata shape: {adata.shape}')
index = adata.obs['ann_finest_level'].isin(['Unknown'])
adata = adata[~index]
print(f'after remove Unknown, adata shape: {adata.shape}')

results_df = []
df = adata.obs[[
    'ann_finest_level', 'ann_level_1', f'predict_mapped_{ct_col}', 'dataset', f'predict_mapped_{ct_col}_prob',
    'ann_level_3', 
]]
df = df.rename(columns = {
    f'predict_mapped_{ct_col}': 'predictions', 
    f'predict_mapped_{ct_col}_prob': 'prob',
    'dataset': 'data_file_names'
})
df['methods'] = 'UniGeneX'
# results_df_all.append(df)
df = filter_less_ct(df)
index = df['prob'] > 0.5
print(f'filter uncertain {(1 - index).sum()} cells')
df = df.loc[index]
results_df.append(df)
```

```python
rm_files = [
    'HLCA_Schultze_unpubl', 'HLCA_Shalek_2018'
]
df_all = pd.concat(results_df, axis = 0)
df_all = df_all[~df_all['data_file_names'].isin(rm_files)]
# df_all = df_all[df_all['ann_finest_level'] != 'Unknown']
df_all = df_all[df_all[ct_col] != 'Unknown']

df_all
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Group the results DataFrame by 'methods' and 'data_file_names'
grouped = df_all.groupby(['methods', 'data_file_names'])

# Initialize lists to store metrics
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []

for (method, data_file), group in grouped:
    true_labels = group[ct_col]
    predicted_labels = group['predictions']

    # Calculate metrics
    accuracy = accuracy_score(true_labels, predicted_labels)
    precision = precision_score(true_labels, predicted_labels, average='weighted')
    recall = recall_score(true_labels, predicted_labels, average='weighted')
    f1 = f1_score(true_labels, predicted_labels, average='weighted')

    # Append the calculated metrics to the respective lists
    accuracy_scores.append((method, data_file, accuracy))
    precision_scores.append((method, data_file, precision))
    recall_scores.append((method, data_file, recall))
    f1_scores.append((method, data_file, f1))

# Create DataFrames from the calculated metrics
accuracy_df = pd.DataFrame(accuracy_scores, columns=['methods', 'data_file_names', 'accuracy'])
precision_df = pd.DataFrame(precision_scores, columns=['methods', 'data_file_names', 'precision'])
recall_df = pd.DataFrame(recall_scores, columns=['methods', 'data_file_names', 'recall'])
f1_df = pd.DataFrame(f1_scores, columns=['methods', 'data_file_names', 'f1_score'])
```

```python
arr = adata.X.sum(1)
fig, axs = plt.subplots(1, 4, figsize=(8, 4), gridspec_kw={'wspace': .5, 'hspace': 0.2})
for i, df_vis in enumerate([
    accuracy_df,
    precision_df,
    recall_df,
    f1_df,
]):
    sns.boxplot(x='methods', y=df_vis.columns[-1], data=df_vis, ax = axs[i])
    axs[i].set_ylim([-.05, 1.2])

plt.show()
plt.close()
```

```python
def identify_outliers(umap_coords, threshold_q=.99):
    from sklearn.neighbors import NearestNeighbors

    # Compute k nearest neighbors
    nn = NearestNeighbors(n_neighbors=30).fit(umap_coords)
    distances, indices = nn.kneighbors(umap_coords)

    # Calculate the mean distance to neighbors
    mean_distances = np.mean(distances, axis=1)

    threshold = np.quantile(mean_distances, threshold_q)
    outliers_mask = mean_distances > threshold
    return mean_distances, outliers_mask
#############
### examples
#############

# mean_distances, outliers_mask = identify_outliers(atlas_adata_all.obsm['X_umap'])
# if outliers_mask is not None:
#     adata_vis = adata_vis[~outliers_mask]
```
